# Housing Data Estimates
We'll use the [Boston pricing dataset](https://www.tensorflow.org/api_docs/python/tf/keras/datasets/boston_housing/load_data) to play around with some models.
<br /><br />
!!! NOTE !!!

Carnegie Mellon, who maintains this data, has noted that this data has an ethical problem. Refer to the link above for more information

```python
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.model_selection import train_test_split

def plot_history(history):
  pd.DataFrame(history.history).plot()
  plt.xlabel("Epochs")
  plt.ylabel("Loss")

(X_train, y_train), (X_test, y_test) = tf.keras.datasets.boston_housing.load_data(path='boston_housing.npz', test_split=0.2, seed=42)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
tf.random.set_seed(42)
callback = tf.keras.callbacks.EarlyStopping(monitor="loss", patience=3)
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(13, activation="relu"))
model.add(tf.keras.layers.Dense(26, activation="relu"))
model.add(tf.keras.layers.Dense(13, activation="relu"))
model.add(tf.keras.layers.Dense(1))

model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.Adam(),
              metrics=["mae"],
              )

history_1 = model.fit(X_train, y_train, epochs=1000, callbacks=[callback], verbose=0)
```

```python
plt.plot(np.arange(1, len(history_1.history["mae"])+1), history_1.history["mae"])
plt.plot(np.arange(1, len(history_2.history["mae"])+1), history_2.history["mae"])
plt.plot(np.arange(1, len(history_3.history["mae"])+1), history_3.history["mae"])
plt.xlabel("epochs")
plt.ylabel("mae")
```

```python
def test_model():
  tf.random.set_seed(42)
  callback = tf.keras.callbacks.EarlyStopping(monitor="loss", patience=3)
  model = tf.keras.Sequential()
  model.add(tf.keras.layers.Dense(13, activation="relu"))
  model.add(tf.keras.layers.Dense(26, activation="relu"))
  model.add(tf.keras.layers.Dense(13, activation="relu"))
  model.add(tf.keras.layers.Dense(1))

  model.compile(loss=tf.keras.losses.mae,
                optimizer=tf.keras.optimizers.Adam(),
                metrics=["mae"],
                )

  history = model.fit(X_train, y_train, epochs=1000, callbacks=[callback], verbose=0)
  return history
```

```python
history_a = test_model()
history_b = test_model()
history_c = test_model()
plt.plot(np.arange(1, len(history_a.history["mae"])+1), history_a.history["mae"])
plt.plot(np.arange(1, len(history_b.history["mae"])+1), history_b.history["mae"])
plt.plot(np.arange(1, len(history_c.history["mae"])+1), history_c.history["mae"])
plt.xlabel("epochs")
plt.ylabel("mae")
```

```python
def test_model():
  tf.random.set_seed(42)
  callback = tf.keras.callbacks.EarlyStopping(monitor="loss", patience=3)
  model = tf.keras.Sequential()
  model.add(tf.keras.layers.Dense(13, activation="relu"))
  model.add(tf.keras.layers.Dense(26, activation="relu"))
  model.add(tf.keras.layers.Dense(26, activation="relu"))
  model.add(tf.keras.layers.Dense(13, activation="relu"))
  model.add(tf.keras.layers.Dense(1))

  model.compile(loss=tf.keras.losses.mae,
                optimizer=tf.keras.optimizers.Adam(learning_rate=.01),
                metrics=["mae"],
                )

  history = model.fit(X_train, y_train, epochs=1000, callbacks=[callback], verbose=0)
  return history

history_a = test_model()
history_b = test_model()
history_c = test_model()
plt.plot(np.arange(1, len(history_a.history["mae"])+1), history_a.history["mae"])
plt.plot(np.arange(1, len(history_b.history["mae"])+1), history_b.history["mae"])
plt.plot(np.arange(1, len(history_c.history["mae"])+1), history_c.history["mae"])
plt.xlabel("epochs")
plt.ylabel("loss")
```
